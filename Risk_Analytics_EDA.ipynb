### Finding percentage of null values in each column

```python
round(100*inp0.isnull().sum()/len(inp0)).sort_values(ascending = False)
```

#### As we can see many columns have missing values more than 35%, we will drop those columns 

```python
inp0 = inp0.loc[:, round(100*inp0.isnull().mean()) <35]
```

```python
inp0.shape
```

```python
(inp0.isnull().mean())*100
```

### Checking for missing values <=13%

### AMT_Annuity

```python
inp0.AMT_ANNUITY.describe() ### There is high difference between 75th percentile and maximum value. 
```

```python
sns.boxplot(inp0.AMT_ANNUITY)
plt.show()
```

#### There are outliers in the AMT_ANNUITY column due to which we will be imputing it with the median rather than mean which will give an unreliable representation of data for missing values.

```python
inp0.AMT_ANNUITY.isnull().sum()
```

```python
inp0.AMT_ANNUITY.fillna(inp0.AMT_ANNUITY.median(), inplace = True)
```

```python
inp0.AMT_ANNUITY.isnull().sum()
```

### AMT_GOODS_PRICE

```python
inp0.AMT_GOODS_PRICE.describe()/1000
```

```python
sns.boxplot(inp0.AMT_GOODS_PRICE)
plt.show()
```

#### Imputing the median of the data for AMT_GOODS_PRICE missing values. We see outliers as well but it can just be any large goods price as well so will not consider it as an outlier. 

```python
inp0.AMT_GOODS_PRICE.fillna(inp0.AMT_GOODS_PRICE.median(), inplace = True)
```

```python
inp0.AMT_GOODS_PRICE.isnull().sum()
```
